```python
import matplotlib.pyplot as plt
import numpy as np

def draw(board):
    n = len(board)
    b = np.zeros((n, n, 3))
    b += 0.8
    b[ ::2,  ::2] = 1
    b[1::2, 1::2] = 1

    fig, ax = plt.subplots()
    ax.imshow(b, interpolation='nearest')

    for y, x in enumerate(board):
        ax.text(x, y, u'\u2655', size=30, ha='center', va='center')

    ax.set(xticks=[], yticks=[])
    ax.axis('image')

    plt.show()
```

```python
board = [2, 4, 1, 3, -1, -1, -1, -1]
draw(board)
```

```python
def valid(board, row, col):
    n = len(board)
    for i in range(row):
        if board[i] == col:
            return False
        d = row - i
        if board[i] + d == col or board[i] - d == col:
            return False
    return True
```

```python
def nqueensbt(board, row):
    n = len(board)
    if row < n:
        for col in range(n):
            if valid(board, row, col):
                board[row] = col
                if nqueensbt(board, row + 1):
                    return True
    else:
        draw(board)
        return True
```
